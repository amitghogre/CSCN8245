```python
df = pd.read_csv('income_data.csv')
df
```

```python
from scipy.stats import chi2_contingency

# Create a contingency table of race and education
contingency_table = pd.crosstab(df['race'], df['education'])
# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)
# Print the results
print('Chi-square statistic:', chi2)
print('P-value:', p)
```

```python
# Create the indicator feature
df['college_education'] = (df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])) * 1
# Print the first 10 rows of the df
print(df.head(10))
```
