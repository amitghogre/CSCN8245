**1. The relationship between the penguin's bill length and bill depth for each species.** (5 marks)

```python
## TODO: Check your attributes, make a graph using Matplotlib or Seaborn to show the relationship needed

import matplotlib.pyplot as plt
import numpy as np

# Sample data for illustration (replace with your actual data)



df = pd.read_csv("penguins.csv")

# Create separate DataFrames for each species
adelie_data = df[df['species'] == 'Adelie']
chinstrap_data = df[df['species'] == 'Chinstrap']
gentoo_data = df[df['species'] == 'Gentoo']

# Create a scatterplot with all three species
plt.figure(figsize=(8, 6))

# Scatterplot for Adelie penguins
plt.scatter(adelie_data['bill_length_mm'], adelie_data['bill_depth_mm'], c='blue', label='Adelie', marker='o', s=100)

# Scatterplot for Chinstrap penguins
plt.scatter(chinstrap_data['bill_length_mm'], chinstrap_data['bill_depth_mm'], c='green', label='Chinstrap', marker='s', s=100)

# Scatterplot for Gentoo penguins
plt.scatter(gentoo_data['bill_length_mm'], gentoo_data['bill_depth_mm'], c='orange', label='Gentoo', marker='^', s=100)

plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.title('Penguin Bill Length vs. Bill Depth')
plt.legend()

plt.show()
```

 **What do you observe from the graph?** (2 marks)

Its clear how Chinstrap species has moderate bill depth and bill height. Gentoo species has smaller depth where as adelie has the smallest length.

**Why did you choose this type of graph?** (2 marks)

This graph clearly distinguishes the depth and length of each species without any overlapping.

**2. Suitable graph of the mean BMI (BMI not body_mass weight_in_kg/height(m2)) of female and male penguins. (Assume the height is the flipper length)**  (5 marks)


```python
#Your answer here




df = pd.read_csv("penguins.csv")

df['BMI'] = df['body_mass_g'] / ((df['flipper_length_mm'] / 1000) ** 2)  


mean_bmi_by_sex = df.groupby('sex')['BMI'].mean()


plt.figure(figsize=(8, 6))
mean_bmi_by_sex.plot(kind='bar', color=['blue', 'orange'], alpha=0.9)
plt.xlabel('Sex')
plt.ylabel('Mean BMI')
plt.title('Mean BMI of Penguins ')
plt.xticks(rotation=0)
plt.show()

```

 **What do you observe from the graph?** (2 marks)

Your answer here:

*   Its clear that Male has higher BMI as compared to Female. This could be possible because of various features.

**Why did you choose this type of graph?** (2 marks)

Your answer here:

* Bar graph is the best fit to compare this data as it doesnt overlap and clearly shows the difference.

```python
#Your solution here
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
df = pd.read_csv('museum_visitors.csv')

# Create a line graph showing the number of visitors to each museum over time
df.set_index('Date', inplace=True)

# Create a line plot
df.plot(kind='line', figsize=(15, 6))

# Set the plot title and axis labels
plt.title('Museum Record')
plt.xlabel('Date')
plt.ylabel('Number of visitors')


plt.legend()
plt.show()

```
